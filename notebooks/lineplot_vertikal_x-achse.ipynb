```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
data = pd.read_csv("aufmasse_freigabe_datum_anzahl.csv", sep=",")

data
```

```python
# Filter Labels, so nicht lesbar bei vielen Werten

labels = data["freigabe_datum"].copy()

size = len(labels)
for i in range(0,size):
    if i % 4 != 0:
        labels[i] = ""

labels
```

```python
sn.set(font_scale=1.0)
fig, ax = plt.subplots(figsize=(30,10), dpi=100)
ax.set_xticks(range(len(labels)))
ax.set_xticklabels(labels, rotation=50)
sn.lineplot(x="freigabe_datum", y="count(1)", data=data, ax=ax)
```
